```python
import re
import sys
from nltk.tokenize import sent_tokenize, word_tokenize
import operator
from nltk.corpus import stopwords
stops = set(stopwords.words('russian'))
stops.update(['.', ',', '!', ')', '(','-'])
file = open('test.txt', "r")
text = file.read()
file.close()

wordsFiltered = []
words = word_tokenize(text.lower())
for w in words:
    if w not in stops:
        wordsFiltered.append(w)



stats = {}
for word in wordsFiltered:
    stats[word] = stats.get(word, 0) + 1

dict(sorted(stats.items(), key=lambda item: item[1]))
import pandas as pd
df = pd.DataFrame.from_dict(data=stats, orient='index', columns=['–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ'])
df1 = df.reset_index()
df2 = df1.rename(columns={'index': '—Å–ª–æ–≤–æ'})
df2.sort_values('–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ',ascending=False)[:50]
```
