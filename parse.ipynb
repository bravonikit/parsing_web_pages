```python
from selenium import webdriver
import time 
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
url = 'https://www.wildberries.ru/catalog/74558473/feedbacks?imtId=58978073'
```

```python
from bs4.element import Comment
import urllib.request


def tag_visible(element):
    if element.parent.name in ['style', 'script', 'head', 'title', 'meta', '[document]']:
        return False
    if isinstance(element, Comment):
        return False
    return True

def text_from_html(body):
    soup = BeautifulSoup(body, 'lxml')
    texts = soup.find_all('p', 'feedback__text')
    visible_texts = filter(tag_visible, texts)  
    return u" ".join(t.strip() for t in visible_texts)
```

```python
driver = webdriver.Chrome()
try:
    driver.get(url=url)
    time.sleep(3)
    # driver.find_element(By.CLASS_NAME, "comments__list")
    
    last_height = driver.execute_script("return document.body.scrollHeight")
    for i in range(230):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
    print(last_height)
    soup = BeautifulSoup(driver.page_source , 'html.parser')
    # print(soup.find_all("p", itemprop="reviewBody"))

    with open(r"test.txt", "w") as file:
        num_coments = 0 
        for item in soup.find_all("p", itemprop="reviewBody"):
            file.write(item.text + '\n')
            num_coments+=1
    print(num_coments)
    # soup.find_all("p", itemprop="reviewBody").to_csv('answ1.csv')
    time.sleep(2)



    print(driver.execute_script("return document.body.scrollHeight"))
except Exception as ex:
    print(ex)
finally:
    driver.close()
    driver.quit()
```
